```python
def get_country_name(phone_number: str) -> str:
    if type(phone_number) is not str:
        raise TypeError("phone_number must be a string in the form '+XXOOOOOOOOOO'.")
    if not phone_number.startswith("+"):
        raise LookupError("Number doesn't contain a country code (missing leading '+').")
        
    COUNTRY_MAP = {"1": "North American",
                   "20": "Egypt",
                   "212": "Morocco",
                   "216": "Tunisia",
                   "31": "Netherlands",
                   "32": "Belgium",
                   "34": "Spain",
                   "352": "Luxembourg",
                   "41": "Switzerland",
                   "44": "United Kingdom",
                   "66": "Thailand",
                   "55": "Brazil",
                   "598": "Uruguay",
                   }
    
    digits = phone_number[1:]
    if not digits:
        raise LookupError("Number doesn't contain a country code (no digits after '+').")
    if not digits.isdigit():
        raise ValueError("All characters after '+' must be digits.")
    
    if len(phone_number) not in (12, 13, 14):
        raise ValueError("Invalid amount of characters: expected '+' + 1–3 digit code + 10-digit number.")
    
    code_country = None
    local_phone_number = None
    
    # Code +X10 -> 12
    if (len(phone_number) == 12) and (digits[:1] in COUNTRY_MAP):
        code_country = digits[:1]
        local_phone_number = digits[1:]
    # Code +XX10 -> 13
    elif (len(phone_number) == 13) and (digits[:2] in COUNTRY_MAP):
        code_country = digits[:2]
        local_phone_number = digits[2:]
    # Code +XXX10 -> 14
    elif (len(phone_number) == 14) and (digits[:3] in COUNTRY_MAP):
        code_country = digits[:3]
        local_phone_number = digits[3:]
    else:
        preview = digits
        if len(digits) >= 3:
            preview = digits[:3]

        raise LookupError(f"Unknown country code: +{preview}")


    return COUNTRY_MAP[code_country]


print(get_country_name('+59800000000000'))
print(get_country_name('+6100000000000'))
```
