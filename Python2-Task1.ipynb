1. Write a Python function to find the max of any numbers of arguments (don't use the build in function `max()`)

```python
# You work here
def find_max_number(*number: tuple[int]) -> int:
    max_number = list(sorted(number))
    return max_number[-1]

print(find_max_number(1, 1, -10, 2, 3, 45, 3, 89, 0))
```

2. Write a function that computes the volume of a sphere given its radius.

```python
# volume of a sphere is 4 * (pi * (radius ** 3)) / 3
def find_volume_sphere(radius: int) -> float:
    pi = 3.14159
    return (4 * pi * (radius ** 3)) / 3

print(find_volume_sphere(10))
```

3. Write a function that checks whether a number is in a given range (Inclusive of high and low)

```python
def is_in_range(target: int, low: int, high: int) -> str:
    if target >= low and target <= high:
        return "IN RANGE"
    else:
        return "OUT OF RANGE"

print(is_in_range(10, 1, 11))
print(is_in_range(10, 1, 9))
```

4. Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```python
# Your work here
def count_str_letters(stings: str) -> int:
    count_upper_letter = 0
    count_lower_letter = 0
    for character in stings:
        if character.isupper():
            count_upper_letter += 1
        elif character.islower():
            count_lower_letter += 1
    print(f"The number of Upper case letter is {count_upper_letter}")
    print(f"The number of Lower case letter is {count_lower_letter}")

count_str_letters('AAAaaaBBBbbbb')
```

5. Write a Python function that takes a list and returns a new list with unique elements of the first list. Please use list comprehension as well

```python
# Your work here
def give_new_unique_list(old_list: list) -> list:
    old_list = set(old_list)
    return [element for element in old_list]

print(give_new_unique_list([1, 2, 3, 4, 5, 5]))
print(give_new_unique_list(["a","b","a","c","b"]))
```

6. Write a Python function to check whether a string is pangram or not.

Note : Pangrams are words or sentences containing every letter of the alphabet at least once.
For example : "The quick brown fox jumps over the lazy dog"


```python
# Your work here
def check_pangram(sentence: str) -> str:
    english_alphabet = ['a', 'b', 'c',
                        'd', 'e', 'f',
                        'g', 'h', 'i',
                        'j', 'k', 'l',
                        'm', 'n', 'o',
                        'p', 'q', 'r',
                        's', 't', 'u',
                        'v', 'w', 'x',
                        'y', 'z'
                        ]
    is_pangram = "PANGRAM"
    for character in english_alphabet:
        if character not in sentence:
            is_pangram = "NOT PANGRAM"
            return is_pangram
    return is_pangram
print(check_pangram('The quick brown fox jumps over the lazy dog'))
print(check_pangram("Pack my box with five dozen liquor jugs"))
print(check_pangram("How vexingly quick daft zebras jump"))
print(check_pangram("The five boxing wizards jump quickly"))
print(check_pangram( "The quick brown fox jumps over the lazy dog" ))
print(check_pangram('abcdefghijklmnopqrstuvwxyz'))
```

7. Write a Python function to calculate the fibonacci number of index i (starts with 1). The function accepts the number as an argument and return a single number.
The fibonacci sequence starts with 1,1,2,3,5,8,13,21,.... Therefore, fibonacci of index 6 is number 8.

```python
# Your work here
def fibonacci(number: int) -> int:
    if number == 1:
        return 1
    if number == 0:
        return 0
    return fibonacci(number - 1) + fibonacci(number - 2)

fibonacci(13)
```

8. Create a function which finds the length of each word in the phrase (broken by spaces) and return the values in a list. The function will have an input of a string, and output a list of integers.

```python
# Your work here
def find_length_word(sentence: str) -> list:
    sentence = sentence.strip()
    return list(map(lambda word: len(word), sentence.split(" ")))

find_length_word('asd asdf asdfgsd')
find_length_word('asdf asdf asdf asdf ')
```

9. Write a function that accepts a list of tuples containing the name of the student and their score, sort them by the score from high to low.

```python
# Your work here
# score = [('a', 10), ('b', 11)]

def sort_students_by_score(students: list[tuple[str, int]]) -> list:
    students = students.copy()

    for index1 in range(len(students)):
        for index2 in range(index1 + 1, len(students)):
            if students[index2][1] > students[index1][1]:
                students[index1], students[index2] = students[index2], students[index1]

    return students

sort_students_by_score([('a', 10), ('b', 11), ('c', 30), ('d', 50), ('e', 23)])

```

10. Write a python function that accepts a list of tuples of length 2, where each tuple contains student name and their scores respectively. The function should return the name of the student whose got maximum score. Please use `map()`, and `max()` function is strictly prohibited.

```python
# Your work here
def find_high_score(students_list: list) -> list:
    collect_scores = list(map(lambda score: score[1], students_list))

    highest_score_index = 0
    for index in range(1, len(collect_scores)):
        if collect_scores[index] > collect_scores[highest_score_index]:
            highest_score_index = index

    top_students = []
    for index in range(len(students_list)):
        if students_list[highest_score_index][1] == students_list[index][1]:
            top_students.append(students_list[index][0])
    
    return top_students

find_high_score([('a', 10), ('b', 20), ('c', 30), ('d', 50), ('e', 23), ('f', 50), ('g', 50)])
```
